```python
from bokeh.io import output_file, show, output_notebook, save
from scripts import categoral_group_vbar
from bokeh.layouts import column, row

import pandas as pd
import mlflow
import numpy as np
%load_ext autoreload
%autoreload 2
```

# IDRID Leaderboard

```python
source = {'Team':['VRT', 'PATech', 'iFLYTEK-MIG', 'SOONER'],
        'Microaneurysms':[0.4951, 0.474, 0.5017, 0.4003],
        'Hemorrhages':[0.6804, 0.649, 0.5588, 0.5395],
        'Cotton_Wool_Spot':[0.6995, '-', 0.6588, 0.5369],
        'Exudates':[0.7127, 0.885, 0.8741, 0.739],
        }
df_leaderboard = pd.DataFrame(source)
df_leaderboard.set_index('Team')
```

```python
savepoint = 'http://localhost:5010'
exp_name = 'FundusSegmentation'
mlflow.set_tracking_uri(savepoint)
df = mlflow.search_runs('10')

def filter_d(run):
    return run['tags.mlflow.runName'].split('IDRID')[0]+'IDRID'

df['Datasets'] = df[['tags.mlflow.runName']].apply(filter_d, axis=1)

df
```

```python
idrid_columns = ['tags.mlflow.runName', 'Datasets']+[col for col in df.columns if 'roc' in col and 'retles' not in col and 'FGADR' not in col]+['params.architecture']

idrid_score = df[idrid_columns]
cols =  np.asarray(['Run', 'Datasets', 'Microaneurysms', 'Exudates', 'Cotton_Wool_Spot', 'Hemorrhages', 'Architecture'])
idrid_score.columns = cols
idrid_score = idrid_score[cols[[0, 1, -1, 2,-2,-3,3]]] # Reordering the columns

df_id = idrid_score.dropna()[::-1]
df_id
```

```python
lesions = ['Microaneurysms', 'Hemorrhages', 'Cotton_Wool_Spot', 'Exudates']
output_notebook()

compared = 'Architecture'
group = 'Datasets'
output_file('IDRID_test_set_perf.html')
p1 = categoral_group_vbar(df_id, group, compared, lesions[0], 'ROC', 350, 500)
p2 = categoral_group_vbar(df_id, group, compared, lesions[1], 'ROC', 350, 500)
p3 = categoral_group_vbar(df_id, group, compared, lesions[2], 'ROC', 350, 500)
p4 = categoral_group_vbar(df_id, group, compared, lesions[3], 'ROC', 350, 500)

save(column(row(p1, p2), row(p3,p4)))
```

```python
compared_df = df_leaderboard 
for l in lesions:
    series = df_id.iloc[df_id[l].argmax()]
    df_tmp = pd.DataFrame([series[lesions].array], columns=lesions)
    if series.Run not in compared_df['Team'].array:
        df_tmp['Team'] = series.Run
        compared_df = compared_df.append(df_tmp)
compared_df = compared_df.set_index('Team')
compared_df.to_html('leaderboard_idrid.html')
```

```python
retles_columns = ['tags.mlflow.runName', 'Datasets']+[col for col in df.columns if 'roc' in col and 'retles' in col]+['params.architecture']

retles_score = df[retles_columns]
cols =  np.asarray(['Run', 'Datasets', 'Microaneurysms', 'Exudates', 'Cotton_Wool_Spot', 'Hemorrhages', 'Architecture'])
retles_score.columns = cols
retles_score = retles_score[cols[[0, 1, -1, 2,-2,-3,3]]] # Reordering the columns

df_retles = retles_score.dropna()[::-1]
df_retles = df_retles[df_retles.Run!='C2F-Unet']
df_retles
```

```python
lesions = ['Microaneurysms', 'Hemorrhages', 'Cotton_Wool_Spot', 'Exudates']
output_notebook()

compared = 'Architecture'
group = 'Datasets'
output_file('RetLes_test_set_perf.html')
p1 = categoral_group_vbar(df_retles, group, compared, lesions[0], 'ROC', 350, 500)
p2 = categoral_group_vbar(df_retles, group, compared, lesions[1], 'ROC', 350, 500)
p3 = categoral_group_vbar(df_retles, group, compared, lesions[2], 'ROC', 350, 500)
p4 = categoral_group_vbar(df_retles, group, compared, lesions[3], 'ROC', 350, 500)

save(column(row(p1, p2), row(p3,p4)))
```

```python
compared_df = pd.DataFrame() 
for l in lesions:
    series = df_retles.iloc[df_retles[l].argmax()]
    df_tmp = pd.DataFrame([series[lesions].array], columns=lesions)
    if not 'Model' in compared_df:
        df_tmp['Model'] = series.Run
        compared_df = (df_tmp)
    elif series.Run not in compared_df['Model'].array:
        df_tmp['Model'] = series.Run
        compared_df = compared_df.append(df_tmp)
        
        
compared_df = compared_df.set_index('Model')
compared_df.to_html('leaderboard_retLes.html')
compared_df
```

```python
global_columns = ['tags.mlflow.runName', 'Datasets']+[col for col in df.columns if 'Test_mIoU' in col]+['params.architecture']

global_score = df[global_columns]
cols =  np.asarray(['Run', 'Datasets', 'Test_mIoU_RetLes', 'Test_mIoU_IDRID', 'Architecture'])
global_score.columns = cols

df_global = global_score.dropna()[::-1]
df_global

df_scores = df_global[['Datasets', 'Test_mIoU_IDRID', 'Test_mIoU_RetLes']]
d = df_scores.groupby('Datasets').mean().sort_values('Test_mIoU_RetLes')

from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.palettes import Spectral5

output_notebook()
source = ColumnDataSource(data=dict(
    x=np.asarray(d.index),
    y1=np.asarray(d['Test_mIoU_IDRID']),
    y2=np.asarray(d['Test_mIoU_RetLes']),
))

TOOLTIPS = [
    ("mIoU", "$y"),
]
p = figure(plot_width=600, plot_height=300, y_range=(0.5,0.75), x_range=np.asarray(d.index), title='Average (across models) mIoUs \
calculated for different training sets', tooltips=TOOLTIPS)
p.line('x', 'y1', source=source, width=0.9, line_width=2, legend_label='mIoU IDRID test',  line_color=Spectral5[4], muted_alpha=0.2)
p.line('x', 'y2', source=source, width=0.9, line_width=2, muted_alpha=0.2, legend_label='mIoU RetLes test', line_color=Spectral5[0])
p.legend.location = "center_right"
p.legend.click_policy='mute'
output_file('Summary_mIoUs.html')
save(p)
d
```
